```python
import pandas as pd

# Load Files
Acquisitions = pd.read_csv("./clean_data_v1/clean_acquisitions_v1.csv")
Degrees = pd.read_csv("./clean_data_v1/clean_degrees_v1.csv")
FundingRounds = pd.read_csv("./clean_data_v1/clean_fr_v2.csv")
Funds = pd.read_csv("./clean_data_v1/clean_funds_v1.csv")
Investments = pd.read_csv("./clean_data_v1/clean_investments_v1.csv")
Ipos = pd.read_csv("./clean_data_v1/clean_ipos_v1.csv")
Milestones = pd.read_csv("./clean_data_v1/clean_milestons_v1.csv")
Objects = pd.read_csv("./clean_data_v1/clean_objects_v1.csv")
People = pd.read_csv("./clean_data_v1/clean_people_v1.csv")
Offices = pd.read_csv("./clean_data_v1/clean_offices_v1.csv")
Relationships = pd.read_csv("./clean_data_v1/clean_relationships_v1.csv")

print("="*60)
print("Dataset 로드 완료!")
print("="*60)
```

## 상장된 스타트업 테이블 merge

ipos: 1,259 rows x 10 cols

Acquisitions: 9,562 rows x 8 cols

FundingRounds: 52,928 rows x 12 cols

Funds: 1,564 rows x 9 cols

Investments: 80,902 rows x 3 cols

Objects: 462,620 rows x 41 cols

```python
# ipo ←> objects ←> funding_rounds merge
ipos_company = (
    Ipos
    .merge(Objects, how='left', left_on='ipos_c_id', right_on='objects_cfpr_id')
    .merge(FundingRounds, how='left', left_on='objects_cfpr_id', right_on='fr_c_id')
)
print(f"df_1 행수:{ipos_company.shape} 1259 -> 2054")
```

```python
ipos_company
```

```python
# csv 쓰기 (데이터 프레임 편집, 분석한 데이터 프레임 저장)
ipos_company.to_csv('ipos_company.csv', index=False)
print("전처리된 데이터를 ames_processed.csv 파일로 저장했습니다.")
```

## 인수된 스타트업 merge

```python
# acquisitions ←> objects ←> funding_rounds merge
acquisitions_company = (
    Acquisitions
    .merge(Objects, how='left', left_on='acquired_c_id', right_on='objects_cfpr_id')
    .merge(FundingRounds, how='left', left_on='objects_cfpr_id', right_on='fr_c_id')
)
print(f"df_1 행수:{acquisitions_company.shape} 9562 -> 11933")
```

```python
acquisitions_company
```

```python
# csv 쓰기 (데이터 프레임 편집, 분석한 데이터 프레임 저장)
acquisitions_company.to_csv('acqusitions_company.csv', index=False)
print("전처리된 데이터를 ames_processed.csv 파일로 저장했습니다.")
```

## 스타트업의 성장

```python
# funding_rounds ←> investments ←> objects merge
fr_company = (
    FundingRounds
    .merge(Investments, how='left', on= 'funding_round_id')
    .merge(Objects, how='left', left_on='invested_c_id', right_on='objects_cfpr_id')
)
print(f"df_1 행수:{fr_company.shape} 52928 -> 100642")
```

```python
fr_company
```

```python
# csv 쓰기 (데이터 프레임 편집, 분석한 데이터 프레임 저장)
fr_company.to_csv('fr_company.csv', index=False)
print("전처리된 데이터를 ames_processed.csv 파일로 저장했습니다.")
```
