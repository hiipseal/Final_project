```python
def cal_rate(code):
    if code=="AUD":
        cond1 = (usd_aud["observation_date"]>="2013-03-13") & (usd_aud["observation_date"]<="2013-11-01")
        result = usd_aud.loc[cond1, "DEXUSAL"].mean()

    if code=="CAD":
        cond2 = (usd_cad["observation_date"]>="2006-04-01") & (usd_cad["observation_date"]<="2013-11-12")
        result = usd_cad.loc[cond2, "DEXCAUS"].mean()

    if code=="EUR":
        cond3 = (usd_eur["observation_date"]>="2004-12-31") & (usd_eur["observation_date"]<="2013-12-09")
        result = usd_eur.loc[cond3, "DEXUSEU"].mean()
    

    if code=="GBP":
        cond4 = (usd_gbp["observation_date"]>="2004-12-31") & (usd_gbp["observation_date"]<="2013-12-10")
        result = usd_gbp.loc[cond4, "DEXUSUK"].mean()

    if code=="JPY":
        cond5 = (usd_jpy["observation_date"]>="2004-12-31") & (usd_jpy["observation_date"]<="2011-07-26")
        result = usd_jpy.loc[cond5, "DEXJPUS"].mean()

    if code=="SEK":
        cond6 = (usd_sek["observation_date"]>="2013-10-08") & (usd_sek["observation_date"]<="2013-10-08")
        result = usd_sek.loc[cond6, "DEXSDUS"].mean()
    
    return result
```

```python
cond1 = (df_funds["raised_currency_code"] == "AUD")

df_funds.loc[cond1, ["raised_amount", "raised_currency_code"]]
```

```python
aud_mean = cal_rate("AUD")
cad_mean = cal_rate("CAD")
eur_mean = cal_rate("EUR")
gbp_mean = cal_rate("GBP")
jpy_mean = cal_rate("JPY")
sek_mean = cal_rate("SEK")
usd_mean = 1.0

df_funds["raised_amount_usd"] = np.nan
df_funds["currency_rate"] = np.nan

avg_rate = {
    "AUD": aud_mean,   # 예: 0.75  (U.S. $ to 1 AUD 이면 곱셈)
    "CAD": cad_mean,   # 예: 0.85  (U.S. $ to 1 CAD 이면 곱셈)
    "EUR": eur_mean,
    "GBP": gbp_mean,
    "JPY": jpy_mean,   # 예: 110.0 (JPY to 1 USD 이면 나눗셈)
    "SEK": sek_mean,
    "USD": 1.0         # 이미 USD인 경우 그대로 사용
}

for cur, r in avg_rate.items():
    cond = df_funds["raised_currency_code"] == cur
    
    # 방향에 따라 곱하거나 나누기
    if cur in ["AUD", "CAD", "EUR", "GBP"]:   # "U.S. $ to 1 통화" 타입
        df_funds.loc[cond, "raised_amount_usd"] = (
            df_funds.loc[cond, "raised_amount"] * r
        )
        df_funds.loc[cond, "currency_rate"] = r
    elif cur in ["JPY", "SEK"]:               # "통화 to 1 U.S. $" 타입
        df_funds.loc[cond, "raised_amount_usd"] = (
            df_funds.loc[cond, "raised_amount"] / r
        )
        df_funds.loc[cond, "currency_rate"] = r
    else:  # USD 등
        df_funds.loc[cond, "raised_amount_usd"] = df_funds.loc[cond, "raised_amount"]
        df_funds.loc[cond, "currency_rate"] = r


#예시 aud
display(df_funds.loc[cond1, ["raised_currency_code", "currency_rate", "raised_amount", "raised_amount_usd"]])
```
