### birthplace 정규화 및 카테고리 분류 함수

```python
import re

# 1. birthplace → 나라 단위로 매핑하는 함수
def birthplace_to_country(x):
    if pd.isna(x):
        return "Missing"
    s = str(x).strip().lower()

    # 미국 표현 통일
    s = s.replace("u.s.a", "usa").replace("u.s.", "usa")

    country_patterns = {
        "United States": [
            "usa", "united states",
            "california", "new york", "los angeles",
            "chicago", "boston", "san francisco", "san diego",
            "seattle", "washington, dc", "washington dc",
            "silicon valley", "palo alto", "mountain view",
            "redwood city", "menlo park", "sunnyvale",
            "santa clara", "san jose", "oakland", "berkeley",
            "austin", "dallas", "houston", "texas", "florida",
            "miami", "orlando", "atlanta", "georgia", "colorado",
            "denver", "arizona", "phoenix", "new jersey", "nj",
            "philadelphia", "pennsylvania", "baltimore", "maryland",
            "virginia", "kansas city", "las vegas", "nevada",
            "ohio", "cleveland", "minnesota", "minneapolis",
            "wisconsin", "madison", "portland, or", "oregon",
            "boulder", "san antonio"
        ],
        "United Kingdom": [
            "united kingdom", "uk", "england", "london",
            "manchester", "cambridge", "oxford", "edinburgh",
            "glasgow", "scotland", "wales", "surrey",
            "brighton", "bristol", "birmingham"
        ],
        "India": [
            "india", "delhi", "new delhi", "bangalore",
            "bengaluru", "chennai", "mumbai", "bombay",
            "pune", "hyderabad", "kolkata", "calcutta"
        ],
        "Canada": [
            "canada", "toronto", "montreal", "vancouver",
            "ottawa", "calgary", "edmonton", "waterloo"
        ],
        "France": [
            "france", "paris", "lyon", "marseille"
        ],
        "Germany": [
            "germany", "berlin", "munich", "hamburg", "frankfurt"
        ],
        "Israel": [
            "israel", "tel aviv", "haifa", "jerusalem"
        ],
        "Australia": [
            "australia", "sydney", "melbourne", "brisbane", "perth"
        ],
        "Brazil": [
            "brazil", "sao paulo", "rio de janeiro"
        ],
        "China": [
            "china", "beijing", "shanghai", "hong kong", "shenzhen"
        ],
        "Japan": [
            "japan", "tokyo", "osaka", "kyoto"
        ],
        "South Africa": [
            "south africa", "johannesburg", "cape town"
        ],
        "Singapore": [
            "singapore"
        ],
        "Russia": [
            "russia", "moscow", "saint petersburg", "st. petersburg"
        ],
        "Spain": [
            "spain", "barcelona", "madrid"
        ],
        "Italy": [
            "italy", "rome", "milan"
        ],
        "Ireland": [
            "ireland", "dublin"
        ],
        "Sweden": [
            "sweden", "stockholm"
        ],
        "Denmark": [
            "denmark", "copenhagen"
        ],
        "Netherlands": [
            "netherlands", "amsterdam"
        ],
        "Turkey": [
            "turkey", "istanbul"
        ],
        "Korea": [
            "korea", "south korea", "seoul"
        ],
        "Pakistan": [
            "pakistan", "karachi", "lahore"
        ],
        "Bangladesh": [
            "bangladesh", "dhaka"
        ],
        "Argentina": [
            "argentina", "buenos aires"
        ],
        "Mexico": [
            "mexico", "mexico city"
        ],
        "Malaysia": [
            "malaysia", "kuala lumpur"
        ],
        "New Zealand": [
            "new zealand", "auckland", "wellington"
        ],
        "Austria": [
            "austria", "vienna"
        ],
        "Switzerland": [
            "switzerland", "zurich", "geneva"
        ],
        "Finland": [
            "finland", "helsinki"
        ],
    }

    # 1) 패턴으로 매칭
    for country, patterns in country_patterns.items():
        for p in patterns:
            if p in s:
                return country

    # 2) 문자열을 깨끗하게 한 뒤, 국가 이름과 완전 일치하면 그 국가로
    simple = re.sub(r"[^a-z ]", " ", s)
    simple = re.sub(r"\s+", " ", simple).strip()
    if simple:
        for country in country_patterns.keys():
            if simple == country.lower():
                return country

    # 3) 그 외는 Other
    return "Other"

# 2. 중간 컬럼: 나라 단위 정규화
copy_people_clean["bplace_country"] = copy_people_clean["birthplace"].apply(birthplace_to_country)

# 3. 결측 플래그: is_people_bplace_missing
copy_people_clean["is_people_bplace_missing"] = copy_people_clean["birthplace"].isna().astype(int)

# 5. 상위 19개 나라 + 나머지 Other 로 cat_people_bplace 생성
top_n = 19
top_labels = copy_people_clean["bplace_country"].value_counts().head(top_n).index.tolist()

def map_country_cat(c):
    return c if c in top_labels else "Other"

copy_people_clean["cat_people_bplace"] = copy_people_clean["bplace_country"].map(map_country_cat)

# 6. 분포 확인
print(copy_people_clean["cat_people_bplace"].value_counts())
```

```python
"""
[copy_milestones(Milestones) 데이터 전처리 함수]
- source_url, source_description, updated_at, created_at 열 삭제
- milestone_code : 모든 값이 other로 동일하여 열 삭제
- milestone_at : 형식 변경 -> datetime
- object_id -> p_object_id 로 열 이름 변경
"""
def Clean_Milestones(copy_milestones):
    # 열 삭제: 'id', 'source_url', 'source_description' 및 기타 3 열
    copy_milestones = copy_milestones.drop(columns=['id', 'source_url', 'source_description', 'created_at', 'updated_at', 'milestone_code'])
    # 열 유형을 datetime64[ns](으)로 변경: 'milestone_at'
    copy_milestones = copy_milestones.astype({'milestone_at': 'datetime64[ns]'})
    # 열 'object_id' 이름을 'mile_cfpr_id'(으)로 변경
    copy_milestones = copy_milestones.rename(columns={'object_id': 'mile_cfpr_id'})
    return copy_milestones

milestones_clean = Clean_Milestones(copy_milestones)
milestones_clean.head()
```

```python
import re

# 1) Text Normalization
def normalize_text(text):
    if pd.isna(text):
        return text
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)  # 특수문자 제거
    text = re.sub(r"\s+", " ", text).strip()  # 다중공백 제거
    return text

milestones_clean["clean_mile_description"] = milestones_clean["description"].apply(normalize_text)

# 2) Event Category Mapping (20개 이하)
event_map = {
    "launch": [
        r"launch", r"release", r"beta", r"open", r"go live", r"rollout", r"introduce"
    ],
    "acquisition": [
        r"acquire", r"acquisition", r"buy", r"merge"
    ],
    "funding": [
        r"raise", r"investment", r"financing", r"funding", r"series [abcde]"
    ],
    "product_update": [
        r"update", r"version", r"deploy", r"upgrade"
    ],
    "hiring": [
        r"hire", r"join", r"appoint", r"recruit"
    ],
    "leadership_change": [
        r"ceo", r"cfo", r"cto", r"resign", r"promot"
    ],
    "partnership": [
        r"partner", r"collaborat"
    ],
    "office_opening": [
        r"office", r"expand location"
    ],
    "ipo": [
        r"ipo", r"public offering"
    ],
    "shutdown": [
        r"shut", r"discontinue", r"close"
    ],
    "rebrand": [
        r"rebrand", r"rename"
    ],
    "campaign": [
        r"campaign", r"marketing launch"
    ],
    "award": [
        r"award", r"recognition"
    ],
    "patent": [
        r"patent"
    ],
    "pricing": [
        r"pricing", r"free plan"
    ],
    "milestone": [
        r"surpass", r"reach", r"users", r"customers"
    ],
    "legal": [
        r"lawsuit", r"bankruptcy"
    ]
}

def extract_event_category(text):
    if pd.isna(text) or text == "":
        return "unknown"
    for category, patterns in event_map.items():
        for p in patterns:
            if re.search(p, text):
                return category
    return "other"

milestones_clean["cat_mile_description"] = milestones_clean["clean_mile_description"].apply(extract_event_category)

# 결과 확인
print(milestones_clean["cat_mile_description"].value_counts())
print("Unique categories:", milestones_clean["cat_mile_description"].nunique())
```

```python
"""
copy_relationships(Relationships) 테이블 전처리 함수
"""
def clean_data(copy_relationships):
    # 열 삭제: 'id', 'created_at', 'updated_at'
    copy_relationships = copy_relationships.drop(columns=['id', 'created_at', 'updated_at'])
    # 열 유형을 datetime64[ns](으)로 변경: 'start_at', 'end_at'
    copy_relationships = copy_relationships.astype({'start_at': 'datetime64[ns]', 'end_at': 'datetime64[ns]'})
    # 열 'person_object_id' 이름을 'rel_p_id'(으)로 변경
    copy_relationships = copy_relationships.rename(columns={'person_object_id': 'rel_p_id'})
    # 열 'relationship_object_id' 이름을 're_cf_id'(으)로 변경
    copy_relationships = copy_relationships.rename(columns={'relationship_object_id': 'rel_cf_id'})
    # stat_at > end_at인 행 삭제
    copy_relationships = copy_relationships[~(copy_relationships['start_at'] > copy_relationships['end_at'])]
    # end_at 플래그 생성 -> has_end_date : 날짜 있으면 1, 결측이면 0
    copy_relationships["has_end_date"] = copy_relationships["end_at"].notna().astype(int)
    return copy_relationships

copy_relationships_clean = clean_data(copy_relationships.copy())
copy_relationships_clean.head()
```

### title(직책) 정규화 및 카테고리 분류 함수

```python
def Title_Category(copy_relationships_clean):
    import re
    import pandas as pd

    df = copy_relationships_clean.copy()

    df["clean_rel_title"] = (
        df["title"]
        .astype(str)
        .str.lower()
        .str.replace(r"[-/,]", " ", regex=True)
        .str.replace(r"\bat\b.*", "", regex=True)
        .str.replace(r"[^a-z0-9\s&]", " ", regex=True)
        .str.replace(r"\s+", " ", regex=True)
        .str.strip()
    )

    priority_map = [
        ("CEO", ["ceo", "chief executive officer", "chief executive", "chairman", "president"]),
        ("Founders", ["founder", "co founder", "co-founder"]),
        ("C-Level", ["cfo", "coo", "cto", "cio", "cmo", "chief"]),
        ("Board", ["board"]),
        ("VP", ["vp", "vice"]),
        ("Director", ["director"]),
        ("Head", ["head"]),
        ("Product", ["product"]),
        ("Engineering", ["engineer", "developer", "software", "hardware"]),
        ("Research", ["scientist", "research"]),
        ("Design", ["design", "ux", "ui"]),
        ("Marketing", ["marketing", "brand"]),
        ("Sales", ["sales", "business development", "bd"]),
        ("HR / People", ["hr", "recruit", "talent"]),
        ("Finance", ["finance"]),
        ("Legal", ["legal", "attorney", "law"]),
        ("Operations", ["operation", "ops"]),
        ("Advisor", ["advisor", "consultant"]),
        ("Intern", ["intern"]),
        ("Staff", ["associate", "assistant", "analyst"]),
    ]

    def map_title(x):
        text = str(x).lower()
        for category, keywords in priority_map:
            for kw in keywords:
                if kw in text:
                    return category
        return "Other"

    df["cat_rel_title"] = df["clean_rel_title"].apply(map_title)

    print("cat_rel_title 분포:")
    print(df["cat_rel_title"].value_counts())
    print("\nUnique:", df["cat_rel_title"].nunique())

    return df
```

```python
# 함수 실행
copy_relationships_clean = Title_Category(copy_relationships_clean)

# 결과 확인
copy_relationships_clean[["title", "cat_rel_title"]].head()
```
