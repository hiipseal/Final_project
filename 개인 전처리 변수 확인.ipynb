```python
# 0 제외 인수 금액 히스토그램
valid = Acquisitions[Acquisitions['price_amount'] > 0]

plt.figure(figsize=(8, 5))
sns.histplot(np.log(valid['price_amount']), bins=20, color='skyblue')
plt.title('인수 금액')
plt.show()
```

```python
# 0 제외 박스플롯
sns.boxplot(x=np.log(valid['price_amount']))
plt.title('인수액 분포')
plt.show()
```

```python
# 인수 날짜별 금액 비교
df = valid.copy()
df['year'] = pd.to_datetime(df['acquired_at']).dt.year
df['log_price'] = np.log(df['price_amount'])

sns.lineplot(data=df, x='year', y='log_price')
plt.title("연도별 인수 금액")
plt.ylabel("price_amount")
plt.show()
```

```python
# 날짜별 인수 건수 비교
df = valid.copy()
df['year'] = pd.to_datetime(df['acquired_at']).dt.year

sns.countplot(data=df, x='year')
plt.title("연도별 인수 금액")
plt.xlabel("연도")
plt.ylabel("인수 건수")
plt.xticks(rotation=45)
plt.show()
```

```python
# price_currency_code 어디 화폐가 많은지

currency_counts = Acquisitions['price_currency_code'].value_counts(dropna=False)
currency_counts


plt.figure(figsize=(8, 5))
sns.barplot(x=currency_counts.index, y=currency_counts.values, palette='pastel')
plt.title('화폐 분포')
plt.show()
```
